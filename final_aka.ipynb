The first three columns are year, month, and day. The remaining 12columns are average windspeeds in knots at 12 locations in Ireland on that day.

Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

```python
wind = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data", delim_whitespace=True)
wind.head()      
```

```python
def func(val):
    if val < 10:
        return(str(0)+str(val))
    else:
        return(str(val))

wind["date"] = wind[["Yr", "Mo", "Dy"]].apply(lambda x: str(x["Yr"])+"-"+func(x["Mo"])+"-"+func(x["Dy"]), axis=1 )
wind["date"] = pd.to_datetime(wind["date"], format="%y-%m-%d")
wind.head()
```

```python
#No, 2061 didn't come yet so I have changed it in 1961
wind["date"] = wind[["Yr", "Mo", "Dy"]].apply(lambda x: "19"+str(x["Yr"])+"-"+func(x["Mo"])+"-"+func(x["Dy"]), axis=1 )
wind["date"] = pd.to_datetime(wind["date"])
wind.head()
```

```python
wind.drop(["Yr", "Mo", "Dy"], axis=1, inplace=True)
wind.set_index("date" , inplace=True)
wind.head()
```

```python
wind.isnull().sum()
```

```python
print(wind.info())
print()
print(wind.notnull().sum())
```

```python
wind.mean(axis=1).mean(axis=0)
```

```python
loc_stats = pd.DataFrame()
loc_stats["min"] = wind.min(axis=0)
loc_stats["max"] = wind.max(axis=0)
loc_stats["mean"] = wind.mean(axis=0)
loc_stats["std"] = wind.std(axis=0)
loc_stats.head()
```

```python
day_stats = pd.DataFrame()
day_stats["min"] = wind.min(axis=1)
day_stats["max"] = wind.max(axis=1)
day_stats["mean"] = wind.mean(axis=1)
day_stats["std"] = wind.std(axis=1)
day_stats.head()
```

```python

import seaborn as sb
```

```python
remuneration = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
```

```python

x = remuneration["Total Arcade Revenue (billions)"]

y = remuneration["Computer Science Doctorates Awarded (US)"]


year = remuneration.Year
colours = ['blue','brown','yellow','gold','red','purple','black','white','green','black']


fig, ax = plt.subplots(figsize=(18,10))


scatter = plt.scatter(x, y, alpha=0.5, label=year, c = colours, s =500)

for i, txt in enumerate(year):
    ax.annotate(txt, (x[i], y[i]))
    
plt.xlabel('Total Revenue earned by arcades in billion)', fontsize = 14)
plt.ylabel('Total Computer Science Doctorates Awarded', fontsize = 14)
plt.title(' the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009', fontsize = 14)
plt.show()
```
